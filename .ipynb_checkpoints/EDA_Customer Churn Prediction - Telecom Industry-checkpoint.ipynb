### 3. Datatype Correction 

```python
df.dtypes
```

- Here, we can see that the datatype of `TotalCharges` is `Object` which is not correct, It should be `float64` because in the dataset it is given like that.

```python
# correcting datatype 
'''we are using to_numeric because when invalid value is encounter it
will ignore it and still covert the feature in float data type '''

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce') #'coerce'=True then invalid parsing will be set as NaN.
df.dtypes
```

```python
# now checking for NaN values 
df.isnull().sum()
```

- There is a NaN values in the `TotalCharges` feature 

```python
# Let's remove them 
df.dropna(how='any',inplace=True)
df.isnull().sum()
```

### 4. Coverting the tenure into bins of 12 months 

```python
df.tenure.value_counts()
# the distribution of tenure is not good for finding insights, so we are going to normalize it by binning the data into 12 months
```

```python
# maximum tenure of the customers
df.tenure.max()
```

```python
l = ["{0} - {1}".format(i, i+11) for i in range(1,72,12)]
l
```

```python
# creating bins of 12 months 
l = ["{0} - {1}".format(i, i+11) for i in range(1,72,12)]
df["tenure_group"] = pd.cut(df.tenure,bins= range(1,80,12),labels=l,right=False) # "right" help in including "1" tenure in bin.
# pd.cut is used to Bin values into discrete intervals.
df
```

```python
df['tenure_group'].value_counts()
```

### 5. Removing Unwanted Features

- So, now we have done the binning of the `tenure` and created a new feature `tenure_group` so there is no need for `tenure` feature anymore.
- And `customerID` is also not needed for the finding insights.

```python
df.drop(columns=['customerID','tenure'],axis=1,inplace=True)
```

#### 1. Disribution of all features on the basis of Churning

```python
columns = df_explore.drop(columns=['Churn','MonthlyCharges','TotalCharges']).columns # not including these columns

plt.figure(figsize=(20,20))
plot = 1 
for col in columns: 
    if plot <= len(columns):
        plt.subplot(5,4,plot)
        sns.countplot(data=df_explore,x=col,hue='Churn')
        plt.title(f"{col} vs Churn",fontsize=15)
        plt.xlabel(col)
    plot+=1
plt.tight_layout()
plt.show()
```

<p style="color:red; font-size:30px;">Insights</p>

1. Gender vs Churn
   - Distribution of the `male` and `female` in term of churning is nearly identical.
   - Hence, we can not say that churning is depending on the gender for now!.

2. SenoirCitizen vs Churn

3. Partner Vs Churn
    - Partner indicates that the customer is married or not .
    - Again the distribution of the `partner` in term of churning is nearly identical.
4. Dependents Vs Churn
    - Dependents means the customer have family member or person who is financially dependent to the customer.
    - Here, The customer which don't have `Dependents` are more likely to `Churn` than the customer without any `Dependents`.
    - The customer who have `Dependents` are more prone to churn and the reason could be they found better service, price etc.
5. PhoneService Vs Churn
    - Customers with phone service are much more likely to churn than customers without phone service.
6. MultipleLines vs Churn
    - Customers with `multiple lines` or `no multiple lines` are more likely to churn than customers with no phone service.
7. InternetService vs Churn
    - Customers with `fiber optic internet` are more likely to churn than customers with `no fiber optic internet` or `no internet service`, reason can be that the fiber optic internet quality is not good price are very high and internet speed is low.
8. OnlineSecurity vs Churn
    - Customers with `online security` are less likely to churn than customers without online security or no internet service.
    - Customers with `no online security` are most likely to churn than customers with online security or no internet service.
9. OnlineBackup vs Churn
    - Customers with `no online backup` are most likely to churn than customers with online backup or no internet service.
10. DeviceProtection vs Churn
    - Customers `without device protection` are more likely to churn than customers `with device protection` or no internet service.
11. TechSupport vs Churn
    - Customers `without tech support` are more likely to churn than customers with tech support or no internet service.
12. StreamingMovies vs Churn
    - Customers `with streaming movies` are slightly more prone to churn than customers `without streaming movies` or no internet service.
13. Contract vs Churn
    - Customers with `month-to-month contracts` are much more likely to churn than customers with `one-year or two-year contracts`.
14. PaperlessBilling vs Churn
    - Customers `with paperless billing` are more likely to churn than customers `without paperless billing`.
15. PaymentMethod vs Churn
    - Customers `with electronic check payment method` are more likely to churn than `customers with other payment methods`.
16. tenure_group vs Churn
    - Customers with `lower tenure` are more likely to churn than customers with `longer tenure`.

#### 6. Label Encoding the Catagorical features

```python
# copying dataframe in new variable 
df_label = df_explore.copy()
```

```python
label = LabelEncoder()
label
```

```python
df_label.dtypes
```

```python
df.columns[(df.dtypes == 'object')|(df.dtypes == 'int64')|(df.dtypes == 'category')]
```

```python
# using label encoder to encode all the categorical features.
columns = df_label.columns[(df_label.dtypes == 'object')|(df_label.dtypes == 'int64')|(df_label.dtypes == 'category')]
for col in columns: 
    df_label[col] = label.fit_transform(df_label[col])
    Mapping = dict(zip(label.classes_,label.transform(label.classes_)))
    print(f"{col} Mapping:",Mapping)
```

2. TotalCharges Vs Churn
    - TotalCharge is depending on the `MonthlyCharge` paid by the customer and `tenure`.
    - It directly implies that if the `tenure` of customer is high means the customer is using the service for long time hence, `TotalCharges` will also increase it also means that the customer is liking the service or pricing that's why they are using the service, therefore the `Churning` is low when the `TotalCharges` are increasing.

### 3. Checking Distribution of Churn Feature in dataset

```python
plt.figure(figsize=(7,4))
sns.countplot(data=df_label,x=df_label["Churn"],hue=df_label['Churn'])
plt.title("Distribution of Churn",fontsize = 15)
plt.xlabel("Churn",fontsize = 13)
plt.ylabel("Count",fontsize = 13)
plt.show()
```

```python
# Distribution in percentage of Churn Feature 
df_label["Churn"].value_counts()*100/len(df_label['Churn'])
```

<p style="color:red; font-size:30px;">Insights</p>

- As you can see the dataset is `highly Imbalance` so, we have to balance it.
- To balance it we are going to use `SMOTE` (Synthetic Minority Oversampling Technique )

```python
# appling SMOTE 
from imblearn.over_sampling import SMOTEN
balance = SMOTEN(random_state=42)
balance
```

```python
x,y = balance.fit_resample(df_label.drop(columns=["Churn"],axis=1),df_label[["Churn"]])
```

```python
x.shape
```

```python
x
```

```python
y.shape
```

```python
y
```

```python
df_label.columns
```

```python
x.columns
```

```python
# combining DataFrames x and y 
a = np.hstack([x,y])
df_final = pd.DataFrame(a,columns=['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService',
       'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
       'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
       'Contract', 'PaperlessBilling', 'PaymentMethod', 'MonthlyCharges',
       'TotalCharges','tenure_group','Churn'])
df_final
```

```python
sns.countplot(data=df_final,x='Churn',hue='Churn') # now the dataset is Balanced.
```

```python
# exporting dataframe 
df_final.to_csv("tele_churn.csv",index=False)
```
